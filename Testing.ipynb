```python
# Parameters to be set by the user
n = 512 # Size of the image N = n^n
tau = 5 # patch size
rho = 1 # locality parameter
q = 14
lmbda = 0.01
max_iter = 10
use_patches = True
# for the creation of the U orthogonal projector
train_image = load_image("data/flowers.png") 

# Load image
f0 = load_image("data/lena.bmp")

# Display f0
plt.figure(figsize = (5,5))
imageplot(f0)
plt.show()

f = f0

Phi = createPhi(n, 0.2) #second parameter is the value of pixels to keep

u = Phi.dot(f.flatten())

# Display u
plt.figure(figsize = (5,5))
imageplot(u.reshape((n,n)))
plt.show()

            
out = NLRAlgorithm(Phi, u, tau, q, train_image, rho, lmbda, max_iter, use_patches)
f_out = out.main_loop()
ps = psnr(f, f_out)
print("PSNR".center(20))
print(ps)
```
